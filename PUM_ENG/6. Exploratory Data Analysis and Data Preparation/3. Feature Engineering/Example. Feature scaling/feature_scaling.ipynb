```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('weight_height.csv')
```

```python
df.head()
```

```python
sns.regplot(x=df["Height"], y=df["Weight"])
```

# Normalization

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df[['Height']])
df['H_n'] = scaler.transform(df[['Height']])
```

```python
scaler.fit(df[['Weight']])
df['W_n'] = scaler.transform(df[['Weight']])
```

```python
df.head()
```

```python
sns.regplot(x=df["H_n"], y=df["W_n"])
```

# Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['Height']])
df['H_s'] = scaler.transform(df[['Height']])
```

```python
scaler.fit(df[['Weight']])
df['W_s'] = scaler.transform(df[['Weight']])
```

```python
df.head()
```

```python
sns.regplot(x=df["H_s"], y=df["W_s"])
```
