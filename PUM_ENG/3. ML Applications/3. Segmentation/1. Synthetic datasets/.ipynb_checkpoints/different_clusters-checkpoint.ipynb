```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
```

```python
n_samples = 1000
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
```

```python
df_circles = pd.DataFrame(noisy_circles[0], columns =['x', 'y'])
df_moons = pd.DataFrame(noisy_moons[0], columns =['x', 'y'])
df_blobs = pd.DataFrame(blobs[0], columns =['x', 'y'])
```

```python
sns.scatterplot(data = df_circles, x="x", y="y")
```

```python
sns.scatterplot(data = df_moons, x="x", y="y")
```

```python
sns.scatterplot(data = df_blobs, x="x", y="y")
```
